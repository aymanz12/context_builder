```python
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
import chromadb
```

```python
# 1. Définir le modèle d'Embedding (transforme le texte en chiffres)
# all-MiniLM-L6-v2 est rapide, gratuit et très efficace pour les tests
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
# 2. Se connecter au serveur ChromaDB (via Docker)
client = chromadb.HttpClient(host="localhost", port=8000)
```

```python
# 3. Préparer des données sémantiques fictives (Mémoire historique/Habitudes)
documents_contextuels = [
    "L'enfant a l'habitude de jouer dans sa chambre l'après-midi, souvent sur le tapis.",
    "Il y a un historique d'ouvertures de fenêtres non autorisées dans la chambre.",
    "L'enfant s'appelle Léo. Ses parents sont très stricts concernant la zone de la cuisine.",
    "Les pleurs détectés sont souvent liés à la fatigue en fin de journée, vers 15h."
]
```

```python
metadonnées = [
    {"type": "habitude", "pertinence": "haute"},
    {"type": "historique_risque", "pertinence": "critique"},
    {"type": "profil", "pertinence": "moyenne"},
    {"type": "historique_audio", "pertinence": "moyenne"}
]
```

```python
ids = ["doc1", "doc2", "doc3", "doc4"]

print("Génération des embeddings et insertion dans ChromaDB...")

# 4. Créer la collection vectorielle et insérer les données
vector_db = Chroma.from_texts(
    texts=documents_contextuels,
    embedding=embeddings,
    metadatas=metadonnées,
    ids=ids,
    client=client,
    collection_name="kids_semantic_memory"
)

print("Succès ! La mémoire sémantique est prête dans ChromaDB.")
```
