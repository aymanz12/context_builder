```python
from pymongo import MongoClient
from datetime import datetime
```

```python
# 1. Connexion à MongoDB en utilisant vos identifiants Docker
client = MongoClient("mongodb://admin:secret@localhost:27017/")
# 2. Sélection de la base de données 
db = client["context_builder"]
```

```python
# 3. Création/Sélection des collections (équivalent des tables)
alerts_collection = db["alerts"]
events_collection = db["events"]
gps_collection = db["gps_data"]
state_collection = db["child_state"]
```

```python
# Vider les collections avant d'insérer pour éviter les doublons 
# si vous exécutez le script plusieurs fois
alerts_collection.delete_many({})
events_collection.delete_many({})
gps_collection.delete_many({})
state_collection.delete_many({})
```
